Задание 1
Создайте класс Circle (окружность). Для данного
класса реализуйте ряд перегруженных операторов:
■ Проверка на равенство радиусов двух окружностей
(операция = =);
■ Сравнения длин двух окружностей (операции >, <,
<=,>=);
■ Пропорциональное изменение размеров окружности,
путем изменения ее радиуса (операции + - += -=).

```python
import math

class Circle:
    def __init__(self, radius):
        self.radius = radius

    def __eq__(self, other):
        if not isinstance(other, Circle):
            return NotImplemented
        return self.radius == other.radius

    def __gt__(self, other):
        if not isinstance(other, Circle):
            return NotImplemented
        return self.length() > other.length()

    def __lt__(self, other):
        if not isinstance(other, Circle):
            return NotImplemented
        return self.length() < other.length()

    def __le__(self, other):
        if not isinstance(other, Circle):
            return NotImplemented
        return self.length() <= other.length()

    def __ge__(self, other):
        if not isinstance(other, Circle):
            return NotImplemented
        return self.length() >= other.length()

    def __add__(self, other):
        if not isinstance(other, (int, float)):
            return NotImplemented
        return Circle(self.radius + other)

    def __sub__(self, other):
        if not isinstance(other, (int, float)):
            return NotImplemented
        return Circle(self.radius - other)

    def __iadd__(self, other):
        if not isinstance(other, (int, float)):
            return NotImplemented
        self.radius += other
        return self

    def __isub__(self, other):
        if not isinstance(other, (int, float)):
            return NotImplemented
        self.radius -= other
        return self

    def length(self):
        return 2 * math.pi * self.radius

    def __str__(self):
        return f"Circle with radius: {self.radius}"

circle1 = Circle(10)
circle2 = Circle(20)
circle3 = Circle(10)

print(circle1 == circle3)
print(circle1 < circle2)
print(circle1 <= circle3)
print(circle1 > circle2)
print(circle2 >= circle1)

print(f"Length of circle1: {circle1.length()}")
print(f"Length of circle2: {circle2.length()}")

circle1 += 5
print(circle1)

circle1 -= 3
print(circle1)
```

Задание 2
Создайте класс Complex (комплексное число). Более
подробно ознакомиться с комплексными числами можно
по ссылке.
Создайте перегруженные операторы для реализации
арифметических операций для по работе с комплексными
числами (операции +, -, *, /).

```python
class Complex:
    def __init__(self, real=0, imag=0):
        self.real = real
        self.imag = imag

    def __add__(self, other):
        if not isinstance(other, Complex):
            return NotImplemented
        return Complex(self.real + other.real, self.imag + other.imag)

    def __sub__(self, other):
        if not isinstance(other, Complex):
            return NotImplemented
        return Complex(self.real - other.real, self.imag - other.imag)

    def __mul__(self, other):
        if not isinstance(other, Complex):
            return NotImplemented
        real_part = self.real * other.real - self.imag * other.imag
        imag_part = self.real * other.imag + self.imag * other.real
        return Complex(real_part, imag_part)

    def __truediv__(self, other):
        if not isinstance(other, Complex):
            return NotImplemented
        denom = other.real**2 + other.imag**2
        if denom == 0:
            raise ZeroDivisionError("division by zero")
        real_part = (self.real * other.real + self.imag * other.imag) / denom
        imag_part = (self.imag * other.real - self.real * other.imag) / denom
        return Complex(real_part, imag_part)

    def __str__(self):
        return f"{self.real} + {self.imag}i"


c1 = Complex(2, 3)
c2 = Complex(1, 4)

print("c1:", c1)
print("c2:", c2)


c3 = c1 + c2
print("c1 + c2 =", c3)


c4 = c1 - c2
print("c1 - c2 =", c4)


c5 = c1 * c2
print("c1 * c2 =", c5)


c6 = c1 / c2
print("c1 / c2 =", c6)
```

Задание 3
Вам необходимо создать класс Airplane (самолет). С помощью перегрузки операторов реализовать: 
■ Проверка на равенство типов самолетов (операция
= =);
■ Увеличение и уменьшение пассажиров в салоне самолета (операции + - += -=);
■ Сравнение двух самолетов по максимально возможному количеству пассажиров на борту (операции >
< <= >=).

```python
class Airplane:
    def __init__(self, plane_type, max_passengers):
        self.plane_type = plane_type
        self.max_passengers = max_passengers
        self.current_passengers = 0

    def __eq__(self, other):
        if not isinstance(other, Airplane):
            return NotImplemented
        return self.plane_type == other.plane_type

    def __add__(self, number):
        if not isinstance(number, int):
            return NotImplemented
        if self.current_passengers + number > self.max_passengers:
            raise ValueError("Превышено максимальное количество пассажиров!")
        self.current_passengers += number
        return self

    def __sub__(self, number):
        if not isinstance(number, int):
            return NotImplemented
        if self.current_passengers - number < 0:
            raise ValueError("Количество пассажиров не может быть отрицательным!")
        self.current_passengers -= number
        return self

    def __iadd__(self, number):
        return self.__add__(number)

    def __isub__(self, number):
        return self.__sub__(number)

    def __gt__(self, other):
        if not isinstance(other, Airplane):
            return NotImplemented
        return self.max_passengers > other.max_passengers

    def __lt__(self, other):
        if not isinstance(other, Airplane):
            return NotImplemented
        return self.max_passengers < other.max_passengers

    def __le__(self, other):
        if not isinstance(other, Airplane):
            return NotImplemented
        return self.max_passengers <= other.max_passengers

    def __ge__(self, other):
        if not isinstance(other, Airplane):
            return NotImplemented
        return self.max_passengers >= other.max_passengers

    def __str__(self):
        return f"{self.plane_type} - Макс. пассажиров: {self.max_passengers}, Текущие пассажиры: {self.current_passengers}"


plane1 = Airplane("Boeing 747", 400)
plane2 = Airplane("Airbus A380", 500)

print(plane1)
print(plane2)

print(plane1 == plane2)

plane1 += 100
print(plane1)

plane1 -= 50
print(plane1)

print(plane1 < plane2)
print(plane1 >= plane2)

plane2 += 200
print(plane2)
```

Задание 4
Создать класс Flat (квартира). Реализовать перегруженные операторы:
■ Проверка на равенство площадей квартир (операция
==);
■ Проверка на неравенство площадей квартир (операция !=);
■ Сравнение двух квартир по цене (операции > < <= >=).

```python
class Flat:
    def __init__(self, area, price):
        self.area = area
        self.price = price

    def __eq__(self, other):
        """Проверка на равенство площадей квартир."""
        if not isinstance(other, Flat):
            return NotImplemented
        return self.area == other.area

    def __ne__(self, other):
        """Проверка на неравенство площадей квартир."""
        return not self.__eq__(other)

    def __gt__(self, other):
        """Сравнение квартир по цене (больше)."""
        if not isinstance(other, Flat):
            return NotImplemented
        return self.price > other.price

    def __lt__(self, other):
        """Сравнение квартир по цене (меньше)."""
        if not isinstance(other, Flat):
            return NotImplemented
        return self.price < other.price

    def __ge__(self, other):
        """Сравнение квартир по цене (больше или равно)."""
        if not isinstance(other, Flat):
            return NotImplemented
        return self.price >= other.price

    def __le__(self, other):
        """Сравнение квартир по цене (меньше или равно)."""
        if not isinstance(other, Flat):
            return NotImplemented
        return self.price <= other.price

    def __str__(self):
        """Возвращает строковое представление квартиры."""
        return f"Квартира: Площадь: {self.area} м², Цена: {self.price} валют."



flat1 = Flat(50, 100000)
flat2 = Flat(50, 150000)
flat3 = Flat(75, 200000)

print(flat1)
print(flat2)

print(flat1 == flat2)
print(flat1 != flat3)


print(flat1 < flat2)
print(flat2 > flat3)
print(flat1 <= flat2)
print(flat3 >= flat2)
```
